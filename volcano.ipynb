```python
import folium
import pandas
```

```python

map = folium.Map(location=[40.236062, -111.304886],zoom_start=5,tile="stamen terrain")
```

```python
fg = folium.FeatureGroup(name="Volcano Map",)
map.add_child(fg)
```

```python
data = pandas.read_csv('volcanos.txt')
data
```

```python
lat = data['LAT']
lon = data['LON']
name = data['NAME']
status = data['STATUS']
type = data['TYPE']
location = data['LOCATION']
evl = data['ELEV']
```

```python
def color_pro(value):
    if value < 1000:
        return 'green'
    elif value < 3000:
        return 'orange'
    else:
        return 'red'
```

```python
for lt,ln,n,status,type,loc,e in zip(lat,lon,name,status,type,location,evl):
    fg.add_child(folium.Marker(location=[lt,ln], popup=f"Name:{n} \n\n Value: {e} \n\n Location: {loc}\n Status:{status} \n Type: {type}",icon=folium.Icon(color=color_pro(e),icon="cloud")))
map.save('volcano.html')
```
